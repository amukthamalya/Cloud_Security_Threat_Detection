```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
np.random.seed(42)
```

```python
roles = ["Admin", "DevOps", "Analyst", "Intern", "User"]
countries = ["Ireland", "UK", "Germany", "USA", "India", "France"]
resources = ["S3", "EC2", "IAM", "RDS", "CloudTrail", "Lambda"]
login_status_option = ["Success", "Failure"]
mfa_options = ["Yes", "No"]
privilege_options = ["Yes", "No"]
```

```python
num_records = 2000
start_date = datetime.now() - timedelta(days=30)
data = []

for i in range(num_records):
    user_id = f"user_{random.randint(1,150)}"
    role = random.choice(roles)
    country = random.choice(countries)
    resource = random.choice(resources)
    login_status = random.choice(login_status_option)
    mfa_used = random.choice(mfa_options)
    privilege_escalation = random.choice(privilege_options)

    login_time = start_date + timedelta(
        minutes = random.randint(0, 30*24*60)
    )
    ip_address = f"192.168.{random.randint(1,255)}.{random.randint(1,255)}"

    data_downloaded = np.random.exponential(scale = 50)
    session_duration = np.random.exponential(scale = 30)

    data.append([
        user_id,
        role,
        login_time,
        ip_address,
        country,
        login_status,
        mfa_used,
        resource,
        privilege_escalation,
        round(data_downloaded,2),
        round(session_duration,2)
    ])

columns = [
    "user_id",
    "role",
    "login_time",
    "ip_address",
    "country",
    "login_status",
    "mfs_used",
    "resource_accessed",
    "privilege_escalation_attempt",
    "data_downloaded_MB",
    "session_duration_minutes"
]
df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
df["suspicious_login"] = (
    (df["login_status"] == "Failure") &
    (df["mfs_used"] == "No")
)
df["privilege_threat"] = (
    df["privilege_escalation_attempt"] == "Yes"
)
df["data_exfiltration_risk"] = (
    df["data_downloaded_MB"] > 200
)

df.head()
    
    
```

```python
total_suspicious_logins = df["suspicious_login"].sum()
total_privilege_threats = df["privilege_threat"].sum()
total_data_risks = df["data_exfiltration_risk"].sum()

print("Total Suspicious logins:", total_suspicious_logins)
print("Total Privilege Escalation Attempts:",total_privilege_threats)
print("Total Data Exfiltration Risks:", total_data_risks)
```

```python
role_threats = df.groupby("role")[[
    "suspicious_login",
    "privilege_threat",
    "data_exfiltration_risk"
]].sum()
role_threats
```

```python
import matplotlib.pyplot as plt
role_threats["suspicious_login"].plot(kind = "bar")

plt.title("Suspicious Logins by Role")
plt.ylabel("Count")
plt.xlabel("Role")
plt.xticks(rotation = 45)
plt.show()
```

```python
role_threats.plot(kind = "bar")
plt.title("{Security Threats By Role")
plt.ylabel("Count")
plt.xlabel("Role")
plt.xticks(rotation = 45)
plt.show()
```

```python
role_counts = df["role"].value_counts()
role_percentage = role_threats.div(role_counts, axis=0)* 100
role_percentage
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (8,5))
sns.heatmap(role_percentage, annot = True, fmt = "1f", cmap = "Reds")

plt.title("Risk Percentge by Role")
plt.ylabel("Role")
plt.xlabel("Threat Type")
plt.show()
```

Executive Summary

- Interns show the highest privilege escalation risk (~55%)
- Users exhibit the highest suspicious login rate (~27%)
- Data exfiltration risk remains low but present across all roles.
- Elevated-access roles (DevOps, Admin) demonstrate higher privilege-related threats.


Business Impact:

These insights help security teams:
- Prioritize monitoring for high-risk roles
- Strengthen privilege access controls
-  Improve MFA enforcement strategies
-  Detect potential insider threats early
